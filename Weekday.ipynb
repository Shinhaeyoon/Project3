```python
import pandas as pd
import os
import numpy
```

```python
weekday_amsterdam=pd.read_csv("amsterdam_weekdays.csv")
weekday_amsterdam["location"]="Amsterdam"
weekday_amsterdam.head()
```

```python
weekday_athens=pd.read_csv("athens_weekdays.csv")
weekday_athens["location"]="Athens"
```

```python
weekday_barcelona=pd.read_csv("barcelona_weekdays.csv")
weekday_barcelona["location"]="Barcelona"
```

```python
weekday_berlin=pd.read_csv("berlin_weekdays.csv")
weekday_berlin["location"]="Berlin"
```

```python
weekday_budapest=pd.read_csv("budapest_weekdays.csv")
weekday_budapest["location"]="Budapest"
```

```python
weekday_lisbon=pd.read_csv("lisbon_weekdays.csv")
weekday_lisbon["location"]="Lisbon"
```

```python
weekday_london=pd.read_csv("london_weekdays.csv")
weekday_london["location"]="London"
```

```python
weekday_paris=pd.read_csv("paris_weekdays.csv")
weekday_paris["location"]="Paris"
```

```python
weekday_rome=pd.read_csv("rome_weekdays.csv")
weekday_rome["location"]="Rome"
```

```python
weekday_vienna=pd.read_csv("vienna_weekdays.csv")
weekday_vienna["location"]="Vienna"
```

```python
weekday_concat = pd.concat([weekday_amsterdam, weekday_athens, weekday_barcelona,weekday_berlin,weekday_budapest, weekday_lisbon,weekday_london,weekday_paris,weekday_rome, weekday_vienna  ], ignore_index=True)
```

```python
weekday_concat.head()
```

```python
weekday = weekday_concat.drop(['room_shared', 'room_private','Unnamed: 0'], axis=1)
```

```python
weekday.head()
```

```python
weekday.to_csv('weekday.csv', index=False)
```

```python

# Create sqlite database and cursor
conn = sqlite3.connect('airbnb_eu.db')
c = conn.cursor()
# Create the table
c.execute("""CREATE TABLE IF NOT EXISTS airbnb_eu_weekday (
           realSum real,room_type text,person_capacity integer,host_is_superhost text,multi integer,
            biz integer,cleanliness_rating integer,guest_satisfaction_overall integer,bedrooms integer,dist real,
            metro_dist real,attr_index real, attr_index_norm real,rest_index real,rest_index_norm real,lng real,lat real,
            location text
            )""")
conn.commit()

test = conn.execute('SELECT * from airbnb_eu_weekday')
names = [description[0] for description in test.description]
print(names)

```
